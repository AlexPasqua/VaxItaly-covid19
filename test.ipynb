```python
import pandas as pd
import plotly.graph_objects as go
```

```python
lat_long=pd.read_csv('latitude_longitude_centrivaccinazione')
centri = pd.read_csv('/Users/gaetanoantonicchio/Documents/GitHub/covid19-vax-Italy-analysis/covid19-opendata-vaccini/dati/punti-somministrazione-latest.csv')
centri.head()
```

```python
lat_long.rename(columns={"latitudine ":"latitudine"}, inplace=True)
```

```python
centri = centri.join(lat_long[['latitudine','longitudine']])
centri.head()
```

```python
centri.shape
```

```python
import plotly.express as px

fig = px.scatter_mapbox(centri, lat="latitudine", lon="longitudine",
                        color_discrete_sequence=["red"], zoom=3, height=500)
fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```
